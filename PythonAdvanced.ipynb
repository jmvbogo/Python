```python
#CLasses and Objects

#Creating classes with constructor
class Car:
    amount_cars=0
    
    def __init__(self,manufacturer,model,hp):
        self.manufacturer=manufacturer
        self.model=model
        self.hp=hp
        Car.amount_cars+=1
        
    def print_info(self):
        print('Manufacturer:{}, Model:{}, HP:{}'
             .format(self.manufacturer,
                     self.model,
                     self.hp))
        
    def print_car_amount(self):
        print('Amount:{} '.format(Car.amount_cars))

    #Destructor
    def __del__(self):
        print('Car sold')
        Car.amount_cars-=1
        
```

```python
#Inheritance
class Person:
    
    def __init__(self, name, age):
        self.name=name
        self.age=age
        
    def get_older(self,years):
        self.age+=years
        
class Musician(Person):
    
    def __init__(self, name, age, instrument):
        super(Musician,self).__init__(name,age)
        self.instrument=instrument
        
    def print_instrument(self):
        print('I enjoy the {}'
             .format(self.instrument))
    
    def print_info(self):
        print('Name:{}, Age:{}, Instrument:{}'
             .format(self.name,
                     self.age,
                     self.instrument))
        
prsn1=Musician('Jordan',32,'Guitar')

prsn1.print_info()
print(prsn1.name)
prsn1.print_instrument()
prsn1.get_older(1)
prsn1.print_info()
```

```python
#Overwriting methods
class Animal:
    
    def __init__(self,name):
        self.name=name
        
    def make_sound(self):
        print('*Default Sound*')

class Cat(Animal):
    def __init__(self,name):
        super(Cat,self).__init__(name)
        
    def make_sound(self):
        print('{}: MEOW'
              .format(self.name))
        
class Dog(Animal):
    def __init__(self,name):
        super(Dog,self).__init__(name)
        
    def make_sound(self):
        print('{}: WOOF'
              .format(self.name))
        
cat=Cat('Dawg')
dog=Dog('Kat')

cat.make_sound()
dog.make_sound()
```

```python
#Overloading
class Vector():
    
    def __init__(self,x,y):
        self.x=x
        self.y=y
        
    def __str__(self):
        return 'X: %d, Y: %d'%(self.x,self.y)
    
    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)
    
    def __sub__(self,other):
        return Vector(self.x-other.x,self.y-other.y)
    
v1=Vector(2,3)
v2=Vector(7,7)
v3=v1+v2
v4=v1-v2

print(' Vector 1:',v1,'\n',
     'Vector 2:',v2,'\n',
     'Vector 3:',v3,'\n',
     'Vector 4:',v4,'\n')
```

```python
#Synchronizing threads
import threading
import time

x=8192

lock=threading.Lock()

def halve():
    global x,lock
    lock.acquire()
    while(x>1):
        x/=2
        print(x)
        time.sleep(1)
    print('THE END')
    lock.release()
    
def double():
    global x, lock
    lock.acquire()
    while(x<8192):
        x*=2
        print(x)
        time.sleep(1)
    print('THE END')
    lock.release()
    
t1=threading.Thread(target=halve)
t2=threading.Thread(target=double)

t1.start()
t2.start()
```

```python
#Semaphores
import threading
import time

semaphore=threading.BoundedSemaphore(value=5)

def access(thread_number):
    print('{}: Trying to access...'
         .format(thread_number))
    semaphore.acquire()
    print('{}: Access Granted'
         .format(thread_number))
    print('{}: Waiting 5 seconds...'
         .format(thread_number))
    time.sleep(5)
    semaphore.release()
    print('{}: Releasing!'
         .format(thread_number))
    
for thread_number in range(10):
    t=threading.Thread(target=access,
                      args=(thread_number,))
    t.start()
```

```python
#Daemon threads
import threading
import time
import os
from os import remove

path='text.txt'
text=''

with open('text.txt','w') as file:
    file.write('Python is cool!')

def readFile():
    global path, text
    while True:
        with open(path)as file:
            text=file.read()
        time.sleep(3)
        
def printLoop():
    global text
    for x in range(30):
        print(text)
        time.sleep(1)

t1=threading.Thread(target=readFile,daemon=True)
t2=threading.Thread(target=printLoop)

t1.start()
t2.start()
t2.join()


remove('text.txt')
```

```python
#Classes and Tables
import sqlite3

conn=sqlite3.connect('mydata.db')

c=conn.cursor()

c.execute("""SELECT * FROM persons
          WHERE last_name = 'Doe'""")

class Person():
    def __init__(self, first=None,
                last=None, age=None):
        self.first=first
        self.last=last
        self.age=age
        
    def clone_person(self, result):
        self.first=result[0]
        self.last=result[1]
        self.age=result[2]
        
person1=Person()
person1.clone_person(c.fetchone())

print(person1.first)
print(person1.last)
print(person1.age)
```

```python
#Factorial calculation
def factorial(n):
    if n<1:
        return 1
    else:
        number=n*factorial(n-1)
        return number
```

```python
#XML with SAX & processing xml data
import xml.sax

class GroupHandler(xml.sax.ContentHandler):
   
    def startElement(self,name,attrs):
        self.current=name
        if self.current=='person':
            print('---Person---')
            id=attrs['id']
            print('ID: %s'%id)
            
    def endElement(self,name):
        if self.current=='name':
            print('Name: %s'%self.name)
        elif self.current=='age':
            print('Age: %s'%self.age)
        elif self.current=='weight':
            print('Weight: %s'%self.weight)
        elif self.current=='height':
            print('Height: %s'%self.height)
        self.current=''
        
    def characters(self,content):
        if self.current=='name':
            self.name=content
        elif self.current=='age':
            self.age=content
        elif self.current=='weight':
            self.weight=content
        elif self.current=='height':
            self.height=content



parser=xml.sax.make_parser()
handler=GroupHandler()
parser.setContentHandler(handler)
parser.parse('group.xml')
```

```python
#XML with DOM
import xml.dom.minidom

domtree=xml.dom.minidom.parse('group.xml')

group=domtree.documentElement

persons=group.getElementsByTagName('person')

for person in persons:
    print('---Person---')
    if person.hasAttribute('id'):
        print('ID: %s'%person.getAttribute('id'))
    name=person.getElementsByTagName('name')[0]
    age=person.getElementsByTagName('age')[0]
    weight=person.getElementsByTagName('weight')[0]
    height=person.getElementsByTagName('height')[0]
    print('Name: %s'%name.firstChild.data)
    print('Age: %s'%age.firstChild.data)
    print('Weight: %s'%weight.firstChild.data)
    print('Height: %s'%height.firstChild.data)
```

```python
#Manipulationg XML file
import xml.dom

persons=group.getElementsByTagName('person')

persons[0].getElementsByTagName('name')[0].childNodes[0].nodeValue='Jason Fly'

persons[0].setAttribute('id','10')

domtree.writexml(open('group.xml','w'))
```

```python
#Creating new elements
import xml.dom

newperson=domtree.createElement('person')
newperson.setAttribute('id','6')

name=domtree.createElement('name')
name.appendChild(domtree.createTextNode('Paul Smith'))

age=domtree.createElement('age')
age.appendChild(domtree.createTextNode('67'))

weight=domtree.createElement('weight')
weight.appendChild(domtree.createTextNode('115'))

height=domtree.createElement('height')
height.appendChild(domtree.createTextNode('169'))
                                       
newperson.appendChild(name)
newperson.appendChild(age)
newperson.appendChild(weight)
newperson.appendChild(height)

group.appendChild(newperson)
                                       
domtree.writexml(open('group.xml','w'))
```
