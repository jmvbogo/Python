```python
#CLasses and Objects

#Creating classes with constructor
class Car:
    amount_cars=0
    
    def __init__(self,manufacturer,model,hp):
        self.manufacturer=manufacturer
        self.model=model
        self.hp=hp
        Car.amount_cars+=1
        
    def print_info(self):
        print('Manufacturer:{}, Model:{}, HP:{}'
             .format(self.manufacturer,
                     self.model,
                     self.hp))
        
    def print_car_amount(self):
        print('Amount:{} '.format(Car.amount_cars))

    #Destructor
    def __del__(self):
        print('Car sold')
        Car.amount_cars-=1
        
```

```python
#Creating objects
car1=Car('Tesla','Model X', 525)
car2=Car('Wayne Enterprises','Batmobile', 500)
car3=Car('Ford','Focus RS', 350)
car4=Car('Toyots','Supra', 320)



car1.print_info()
del car2
car3.print_info()
car4.print_info()

car1.print_car_amount()


```

```python
#Inheritance
class Person:
    
    def __init__(self, name, age):
        self.name=name
        self.age=age
        
    def get_older(self,years):
        self.age+=years
        
class Musician(Person):
    
    def __init__(self, name, age, instrument):
        super(Musician,self).__init__(name,age)
        self.instrument=instrument
        
    def print_instrument(self):
        print('I enjoy the {}'
             .format(self.instrument))
    
    def print_info(self):
        print('Name:{}, Age:{}, Instrument:{}'
             .format(self.name,
                     self.age,
                     self.instrument))
        
prsn1=Musician('Jordan',32,'Guitar')

prsn1.print_info()
print(prsn1.name)
prsn1.print_instrument()
prsn1.get_older(1)
prsn1.print_info()
```

```python
#Overwriting methods
class Animal:
    
    def __init__(self,name):
        self.name=name
        
    def make_sound(self):
        print('*Default Sound*')

class Cat(Animal):
    def __init__(self,name):
        super(Cat,self).__init__(name)
        
    def make_sound(self):
        print('{}: MEOW'
              .format(self.name))
        
class Dog(Animal):
    def __init__(self,name):
        super(Dog,self).__init__(name)
        
    def make_sound(self):
        print('{}: WOOF'
              .format(self.name))
        
cat=Cat('Dawg')
dog=Dog('Kat')

cat.make_sound()
dog.make_sound()
```

```python
#Overloading
class Vector():
    
    def __init__(self,x,y):
        self.x=x
        self.y=y
        
    def __str__(self):
        return 'X: %d, Y: %d'%(self.x,self.y)
    
    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)
    
    def __sub__(self,other):
        return Vector(self.x-other.x,self.y-other.y)
    
v1=Vector(2,3)
v2=Vector(7,7)
v3=v1+v2
v4=v1-v2

print(' Vector 1:',v1,'\n',
     'Vector 2:',v2,'\n',
     'Vector 3:',v3,'\n',
     'Vector 4:',v4,'\n')
```

```python
#Synchronizing threads
import threading
import time

x=8192

lock=threading.Lock()

def halve():
    global x,lock
    lock.acquire()
    while(x>1):
        x/=2
        print(x)
        time.sleep(1)
    print('THE END')
    lock.release()
    
def double():
    global x, lock
    lock.acquire()
    while(x<8192):
        x*=2
        print(x)
        time.sleep(1)
    print('THE END')
    lock.release()
    
t1=threading.Thread(target=halve)
t2=threading.Thread(target=double)

t1.start()
t2.start()
```

```python
#Semaphores
import threading
import time

semaphore=threading.BoundedSemaphore(value=5)

def access(thread_number):
    print('{}: Trying to access...'
         .format(thread_number))
    semaphore.acquire()
    print('{}: Access Granted'
         .format(thread_number))
    print('{}: Waiting 5 seconds...'
         .format(thread_number))
    time.sleep(5)
    semaphore.release()
    print('{}: Releasing!'
         .format(thread_number))
    
for thread_number in range(10):
    t=threading.Thread(target=access,
                      args=(thread_number,))
    t.start()
```

```python
#Daemon threads
import threading
import time
import os
from os import remove

path='text.txt'
text=''

with open('text.txt','w') as file:
    file.write('Python is cool!')

def readFile():
    global path, text
    while True:
        with open(path)as file:
            text=file.read()
        time.sleep(3)
        
def printLoop():
    global text
    for x in range(30):
        print(text)
        time.sleep(1)

t1=threading.Thread(target=readFile,daemon=True)
t2=threading.Thread(target=printLoop)

t1.start()
t2.start()
t2.join()


remove('text.txt')
```

```python
#Last in first out queues
import queue

q=queue.LifoQueue()

nmbrs=[1,2,3,4,5,6,7]

for x in nmbrs:
    q.put(x)
    
while not q.empty():
    print(q.get())
```
